## Matplotlib example workflow

```python
# 0: import matplotlib and its dependencies
# 1: prepare the data
x = [1, 2, 4, 5]
y = [10, 20, 30, 40]

# 2: setup plot
fix, ax = plt.subplots(figsize=(5,5)) # figsize = (width, height)

# 3: plot data
ax.plot(x, y)

# 4: customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5: save & show
fix.savefig('sample-plot.png')
```

## Making figures with numpy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# create some data
x = np.linspace(0, 10, 100)
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use same date to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# another scatter
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
mahsolat_prices = {
    "shir": 7000,
    "kare": 5000,
    "mast": 14000
}
fig, ax = plt.subplots()
ax.bar(x=list(mahsolat_prices.keys()), height=mahsolat_prices.values())
ax.set(title="Gheymat mahsolat", ylabel="Gheymat", xlabel="esm mahsool");
```

```python
# Make a plot from a dictionary horizontal
mahsolat_prices = {
    "shir": 7000,
    "kare": 5000,
    "mast": 14000
}
fig, ax = plt.subplots()
ax.barh(list(mahsolat_prices.keys()), mahsolat_prices.values());
```

```python
# make some data and make a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10, 5))

# now we have 4 axis
# we make diffrent figures
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(list(mahsolat_prices.keys()), mahsolat_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplots option2
# subplot option 1
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10, 5))

# now we have 4 axis
# we make diffrent figures
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10),np.random.random(10));
ax[1, 0].bar(list(mahsolat_prices.keys()), mahsolat_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataframe
car_prices = pd.read_csv("cars-exported.csv")
car_prices.head()
```

```python
# first clean some data frame
car_prices["Price"] = car_prices["Price"] // 10
car_prices[:4]
```

```python
car_prices["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_prices))
car_prices[:100]
```

```python
car_prices["Total sale"] = car_prices["Price"].cumsum()
car_prices
```

```python
car_prices.plot(x="Sale Date", y="Total sale");
```

```python
car_prices.plot(x="Year", y="Price", kind="scatter");
```

```python
# NOW GET ON HEART DATASET
```

```python
heart_disease = pd.read_csv("heart.csv")
heart_disease.head()
```

```python
heart_disease.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
# OOP mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 5))
heart_disease.plot(kind="scatter",x="age",y="chol",c="target", ax=ax)
```

```python
# OOP method
fig, ax = plt.subplots(figsize=(10, 6))
# plot the data
scatter = ax.scatter(
        x=heart_disease["age"],
        y=heart_disease["chol"],
        c=heart_disease["target"]);

ax.set(title="Heart disease plot",xlabel="age",ylabel="Cholostrol");

ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(heart_disease["chol"].mean(),linestyle="-")
```
